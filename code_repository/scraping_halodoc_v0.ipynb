```python
# KATEGORI COVID-19

from urllib.request import urlopen
import json
import pandas as pd

for page_number in range(1,4):
    request=urlopen(f'https://www.halodoc.com/api/v1/buy-medicine/categories/covid-19/products?page={page_number}&size=20')
    json_data = request.read().decode('utf-8')
    full_data = json.loads(json_data)
    result = full_data["result"]

    obatlist = []
    
    for i in range (0,len(result)):
        slug = result[i]["slug"]
        new_url="https://www.halodoc.com/api/v1/buy-medicine/products/detail/"
        new_url_slug = new_url+slug

        request=urlopen(new_url_slug)
        json_data = request.read().decode('utf-8')
        full_data = json.loads(json_data)

        nama = full_data["name"]
        min_price = full_data["min_price"]
        base_price = full_data["base_price"]
        selling_unit = full_data["selling_unit"]
        packaging = full_data["packaging"]
        kategori = "Batuk dan Flu" # masih manual
        deskripsi = full_data["description"]
        indikasi_umum = full_data["general_indication"]
        komposisi = full_data["composition"]
        # dosis = full_data["dosage"]
        aturan_pakai = full_data["how_to_use"]
        kontra_indikasi = full_data["contraindication"]
        perhatian = full_data["warning"]
        efek_samping = full_data["side_effects"]
        segmentasi = full_data["segmentation"],
        manufaktur = full_data["manufacturer_name"]
        no_registrasi = full_data["bpom_number"]
        slug = full_data["canon_slug"]
        productID = full_data["external_id"]

        obat = {
            "nama" : nama,
            "min_price" : min_price,
            "base_price" : base_price,
            "selling_unit" : selling_unit,
            "packaging" : packaging,
            "kategori" : kategori,
            "deskripsi" : deskripsi,
            "indikasi_umum" : indikasi_umum,
            "komposisi" : komposisi,
            # "dosis" : dosis,
            "aturan_pakai" : aturan_pakai,
            "kontra_indikasi" : kontra_indikasi,
            "perhatian" : perhatian,
            "efek_samping" : efek_samping,
            "segmentasi" : segmentasi,
            "manufaktur" : manufaktur,
            "no_registrasi" : no_registrasi,
            "slug" : slug,
            "productID": productID
        }  

        obatlist.append(obat)
        
df2 = pd.DataFrame(obatlist)
```
